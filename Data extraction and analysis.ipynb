# Data Extraction

**Function to extract article title and article text**

```python
def extract_article_text(url):
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find and extract the article title
        
        title = soup.title.text.strip()
        
        # Find and extract the article text
        #text = ' '.join([p.get_text() for p in soup.find_all('p')])
        content_elements = soup.findAll(attrs={'class': 'td-post-content'})
        content = '\n'.join(content_element.text.strip() for content_element in content_elements)
        
        
        return title, content
    else:
        
        return None, None
```

```python
for index, row in df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    # Extract article text from the URL
    title, text = extract_article_text(url)
    
    if title and text:
        # Save the extracted article text in a text file
        output_file = f"{url_id}.txt"
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write(title)
            f.write(text)
        print(f"Article text extracted and saved for URL_ID {url_id}")
    else:
        print(f"Failed to extract article text for URL_ID {url_id}")
```

**Extracting Derived variables**

```python
def calculate_score(tokens,positives,negatives):
    positive_score=0
    negative_score=0
    for i in tokens:
        if(i.lower() in positives):
            positive_score+=1
    
        elif(i.lower() in negatives):
            negative_score+=1
    # Calculate Polarity Score
    polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)
    
    # Calculate Subjectivity Score
    total_words = len(tokens)
    subjectivity_score = (positive_score + negative_score) / (total_words + 0.000001)
    return positive_score, negative_score, polarity_score, subjectivity_score
```

# Analysis of Readability


```python
def count_complex_words(words):
    vowels = ['a', 'e', 'i', 'o', 'u']
    complex_word_count = 0

    for word in words:
        vowel_count = sum(1 for char in word if char.lower() in vowels)
        if vowel_count > 2:
            complex_word_count += 1

    return complex_word_count
```

# Syllable Count Per Word


```python
import re
```

```python
vowels=['a','e','i','o','u']
syllable_count=[]

for i in txt_list:
    count=0
    words=word_tokenize(i)
    for word in words:
        x=re.compile('[es|ed]$', re.IGNORECASE)
        if x.search(word):
            # Exclude "es" or "ed" from syllable count
            count += max(1, sum(1 for char in word if char.lower() in vowels) - 1)
        else:
            count += max(1, sum(1 for char in word if char.lower() in vowels))
    syllable_count.append(count)
```
